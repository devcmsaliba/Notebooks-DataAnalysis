## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[['cut', 'price']].groupby('cut').mean().reset_index()
diamonds.head()
```

```
grafico = sns.barplot(data=diamonds, x='cut', y='price', errorbar=None, palette='pastel')
grafico.set(title='Preço do diamante por tipo de corte', xlabel='Corte', ylabel='Preço (USD)');
```

**Insight do gráfico 1**:

Gráfico de barras para representar o preço médio dos diamantes agrupados por tipo de corte

Entendo que o corte Premium é o que mais vale se comparado aos demais, vem seguido do Fair, Very Good, Good e por último e com menos valor de mercado o Ideal.

Além disso, noto que não há uma grande diferença entre os cortes, acredito que isso faça com que o Premium não tenha um grande destaque e tenha uma equiparação aos demais, talvez seja porque os que o compram sejam somente aqueles que de fato entendam e tenham um conhecimento muito aprofundado, o que pode não ser a população no geral, que talvez tenha preferência por outros cortes mais em conta que não aparentam tanta diferença do de ponta.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset('flights')
print(data.to_string())
```

```
# gráfico do exercício 3

flights = data.query('1949 <= year < 1960')

with sns.axes_style ('whitegrid'):

  grafico = sns.lineplot(data=flights, x='month', y='passengers', hue='year', palette='pastel')
  grafico.set(title='Numero de passageiros por mês entre 1949 e 1959', xlabel='Mês', ylabel='Passageiros');
  grafico.get_legend().set_title('Ano');
```

**Insight do gráfico 3**:

Gráfico de linha para representar o número de passageiros por mês entre 1949 e 1959.

Notei um aumento de passageiros no decorrer dos anos que, entendo ser conforme mencionei no exercício anterior, tenha sido um meio de transporte que se popularizou e granhou mais adeptos.

O início do ano nos meses de Janeiro e Fevereiro são as demandas mais baixas, juntamente com Maio e Novembro que são meses que antecedem e procedem os períodos de férias escolares (Julho e Dezembro)

Sendo Julho e Agosto os com maior demanda seguido logo após de Dezembro.
